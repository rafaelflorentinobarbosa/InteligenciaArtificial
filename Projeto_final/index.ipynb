# Poluição global por condados / Municípios
- Resíduos mundiais por países (incluindo todos os tipos de poluição até 2010-2020)

# Resumo do dataset
Dados agregados sobre gestão de resíduos sólidos de todo o mundo. Este banco de dados apresenta as estatísticas coletadas por meio do esforço, cobrindo quase todos os países e mais de 330 cidades. As métricas incluídas cobrem todas as etapas da cadeia de valor da gestão de resíduos, incluindo geração, composição, coleta e descarte de resíduos, bem como informações sobre o usuário

# Informações Relevantes que pretendo usar:

## Geração de Resíduos:
- Total de resíduos gerados por país ou cidade.
- Comparação entre a geração de resíduos em diferentes regiões.

## Composição dos Resíduos:
- Percentuais de diferentes tipos de resíduos (orgânicos, plásticos, papel, metal, etc.).
- Análise da composição ao longo do tempo (se houver dados históricos).

## Taxas de Coleta:
- Percentual de resíduos coletados versus gerados.
- Comparação de taxas de coleta entre países ou regiões.

## Métodos de Descarte:
- Percentual de resíduos descartados em aterros, reciclados, incinerados, etc.
- Análise das práticas de descarte em diferentes contextos.

## Relação com PIB e População:
- Correlação entre PIB de um país e a quantidade de resíduos gerados.
- Comparação entre a geração de resíduos per capita em diferentes países.

## Impacto do Lixo Especial:
- Análise das toneladas de resíduos especiais (como resíduos industriais ou agrícolas) gerados.


# Gráficos Interessantes

## Gráfico de Barras:
- Comparar a quantidade total de resíduos gerados entre diferentes países ou cidades.
- Mostrar a composição dos resíduos (cada tipo em uma barra) para diferentes locais.

## Gráfico de Pizza:
- Visualizar a composição percentual dos resíduos para um único país ou região, destacando os tipos principais.

## Gráfico de Linhas:
- Mostrar a evolução da geração de resíduos ao longo do tempo (se houver dados históricos).
- Comparar a taxa de coleta ao longo dos anos.

## Gráfico de Dispersão (Scatter Plot):
- Analisar a relação entre PIB e a quantidade de resíduos gerados, com pontos representando diferentes países.
- Comparar a geração de resíduos per capita com a população.

## Mapas de Calor:
- Criar um mapa de calor global ou regional mostrando a quantidade de resíduos gerados por país ou cidade, facilitando a visualização de áreas problemáticas.

## Boxplot:
- Analisar a distribuição da quantidade de resíduos gerados em diferentes grupos (por exemplo, países desenvolvidos vs. em desenvolvimento).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Selecionar as colunas relevantes
colunas_desejadas = [
    'country_name',
    'total_msw_total_msw_generated_tons_year'
]

# Criar um novo DataFrame com as colunas selecionadas
dados_lixo = paises_poluentes[colunas_desejadas].copy()

# Remover linhas com valores NaN na coluna de lixo gerado
dados_lixo.dropna(subset=['total_msw_total_msw_generated_tons_year'], inplace=True)

# Ordenar os dados pela quantidade total de lixo gerado
dados_lixo.sort_values(by='total_msw_total_msw_generated_tons_year', ascending=False, inplace=True)

# Selecionar os 10 países que mais produzem lixo
top_paises = dados_lixo.head(10)

# Criar o gráfico
plt.figure(figsize=(12, 6))
plt.barh(top_paises['country_name'], top_paises['total_msw_total_msw_generated_tons_year'], color='skyblue')
plt.xlabel('Total de Lixo Gerado (Toneladas)')
plt.title('Top 10 Países que Mais Produzem Lixo')
plt.gca().invert_yaxis()  # Inverter o eixo y para mostrar o maior primeiro
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Selecionar as colunas relevantes
colunas_residuos = [
    'country_name',
    'total_msw_total_msw_generated_tons_year',
    'composition_food_organic_waste_percent',
    'composition_metal_percent',
    'composition_paper_cardboard_percent',
    'composition_rubber_leather_percent',
    'composition_other_percent',
    'composition_wood_percent',
    'composition_plastic_percent'
]

# Criar um novo DataFrame apenas com as colunas relevantes
dados_residuos = paises_poluentes[colunas_residuos].copy()

# Encontrar os 10 maiores produtores de resíduos
top_10_produtores = dados_residuos.nlargest(10, 'total_msw_total_msw_generated_tons_year')

# Determinar qual tipo de resíduo é o mais desperdiçado em cada país
top_10_produtores['tipo_maximo'] = top_10_produtores.loc[:, 'composition_food_organic_waste_percent':'composition_plastic_percent'].idxmax(axis=1)
top_10_produtores['maximo_valor'] = top_10_produtores.loc[:, 'composition_food_organic_waste_percent':'composition_plastic_percent'].max(axis=1)

# Criar o DataFrame para os resultados
resultados = top_10_produtores[['country_name', 'total_msw_total_msw_generated_tons_year', 'tipo_maximo', 'maximo_valor']]

# Mostrar os resultados
print(resultados)

# Criar um gráfico de barras
plt.figure(figsize=(14, 8))
plt.barh(resultados['country_name'], resultados['total_msw_total_msw_generated_tons_year'], color='lightcoral')
plt.xlabel('Total de Resíduos Gerados (toneladas)')
plt.title('10 Maiores Produtores de Resíduos')
plt.gca().invert_yaxis()  # Inverter o eixo y para mostrar o maior primeiro
plt.xticks(rotation=45)

# Adicionar anotações com o tipo de resíduo
for index, value in enumerate(resultados['total_msw_total_msw_generated_tons_year']):
    plt.text(value, index, f"{resultados['tipo_maximo'].iloc[index]}: {resultados['maximo_valor'].iloc[index]:.1f}%", va='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o dataset
paises_poluentes = pd.read_csv('datasets/global_pollution.csv')

# Substituir valores NaN por 0
paises_poluentes.fillna(0, inplace=True)

# Garantir que as colunas estão no formato numérico
paises_poluentes['gdp'] = pd.to_numeric(paises_poluentes['gdp'], errors='coerce')
paises_poluentes['total_msw_total_msw_generated_tons_year'] = pd.to_numeric(paises_poluentes['total_msw_total_msw_generated_tons_year'], errors='coerce')

# Criar um gráfico
plt.figure(figsize=(12, 6))
plt.scatter(paises_poluentes['gdp'], paises_poluentes['total_msw_total_msw_generated_tons_year'], alpha=0.7)
plt.title('Relação entre PIB e Resíduos Sólidos Urbanos Gerados por País')
plt.xlabel('PIB (em bilhões)')
plt.ylabel('Total de Resíduos Sólidos Urbanos Gerados (toneladas/ano)')
plt.xscale('log')  # Usar escala logarítmica para melhor visualização
plt.yscale('log')  # Usar escala logarítmica para melhor visualização
plt.grid(True)

# Exibir o gráfico
plt.show()
```
