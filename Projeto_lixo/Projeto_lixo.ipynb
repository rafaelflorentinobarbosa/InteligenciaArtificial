## Prever quantas toneladas de lixo o Brasil vai produzir em 2025

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Filtrar apenas os dados do Brasil
df_brasil = df[df["País"] == "Brasil"]

# Selecionar as variáveis independentes e a variável dependente
X = df_brasil[['Ano', 'PIB (US$)', 'População']]  # Incluindo todas as variáveis como uma lista
y = df_brasil['Total Lixo (Ton)']

# Criar e treinar o modelo de regressão linear
model = LinearRegression()
model.fit(X, y)

# Dados de entrada para 2025
X_new = pd.DataFrame({
    'Ano': [2025],
    'PIB (US$)': [2800000000000],
    'População': [217000000]
})

# Fazer a previsão
y_pred = model.predict(X_new)
print("Previsão para 2025:", y_pred[0])
```

```python
# Supondo que os coeficientes sejam:
# Coeficiente do ano: 1000
# Coeficiente do PIB: 0.000001
# Coeficiente da população: 5

# A equação da regressão seria:
# Total Lixo = 1000 * Ano + 0.000001 * PIB + 5 * População + b0

# Interpretação:
# - A cada ano que passa, a produção de lixo aumenta em 1000 toneladas (mantendo PIB e população constantes).
# - Um aumento de 1 dólar no PIB está associado a um aumento de 0.000001 toneladas na produção de lixo (mantendo ano e população constantes).
# - Um aumento de 1 pessoa na população está associado a um aumento de 5 toneladas na produção de lixo (mantendo ano e PIB constantes).
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o dataset
df = pd.read_csv('../datasets/dataset_lixo.csv')

# Filtrar os dados apenas para o Brasil
df_brasil = df[df["País"] == "Brasil"]

# Ordenar o DataFrame por ano (caso não esteja ordenado)
df_brasil = df_brasil.sort_values(by="Ano")

# Calcular a taxa de crescimento anual para lixo total e lixo reciclado
df_brasil['Crescimento Lixo Total (%)'] = df_brasil["Total Lixo (Ton)"].pct_change() * 100
df_brasil['Crescimento Lixo Reciclado (%)'] = df_brasil["Lixo Reciclado (Ton)"].pct_change() * 100

# Calcular a média de crescimento ao longo dos anos
media_crescimento_lixo_total = df_brasil['Crescimento Lixo Total (%)'].mean()
media_crescimento_lixo_reciclado = df_brasil['Crescimento Lixo Reciclado (%)'].mean()

# Obter os valores de 2024 para projetar para 2025
lixo_total_2024 = df_brasil[df_brasil['Ano'] == 2024]["Total Lixo (Ton)"].values[0]
lixo_reciclado_2024 = df_brasil[df_brasil['Ano'] == 2024]["Lixo Reciclado (Ton)"].values[0]

# Prever valores para 2025
lixo_total_2025 = lixo_total_2024 * (1 + media_crescimento_lixo_total / 100)
lixo_reciclado_2025 = lixo_reciclado_2024 * (1 + media_crescimento_lixo_reciclado / 100)

# Adicionar a previsão de 2025 no DataFrame
df_brasil_2025 = pd.DataFrame({
    'País': ['Brasil'],
    'Ano': [2025],
    'Total Lixo (Ton)': [lixo_total_2025],
    'Lixo Reciclado (Ton)': [lixo_reciclado_2025]
})

# Concatenar com o DataFrame original
df_brasil = pd.concat([df_brasil, df_brasil_2025], ignore_index=True)

# Gráfico de barras para comparar o lixo total e reciclado
anos = df_brasil["Ano"]
lixo_total = df_brasil["Total Lixo (Ton)"]
lixo_reciclado = df_brasil["Lixo Reciclado (Ton)"]

# Configurar o gráfico de barras
largura_barra = 0.35  # Largura das barras
indice_anos = range(len(anos))  # Posições no eixo x para os anos

# Criar o gráfico de barras
plt.figure(figsize=(10,6))
plt.bar(indice_anos, lixo_total, width=largura_barra, label='Lixo Total', color='b')
plt.bar([i + largura_barra for i in indice_anos], lixo_reciclado, width=largura_barra, label='Lixo Reciclado', color='g')

# Configurações do gráfico
plt.xlabel('Ano')
plt.ylabel('Quantidade de Lixo (Ton)')
plt.title('Lixo Total e Lixo Reciclado no Brasil por Ano (incluindo previsão para 2025)')
plt.xticks([i + largura_barra/2 for i in indice_anos], anos)  # Definir os anos no eixo x
plt.legend()

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

# Lixo produzido por pessoa X PIB percapita

```python
df = pd.DataFrame(df)

# Filtrar os dados apenas para o Brasil
df_brasil = df.loc[df["País"] == "Brasil"]

# Calcular a média de lixo gerado por pessoa
df_brasil.loc[:, "Lixo por Pessoa (kg)"] = df_brasil["Total Lixo (Ton)"] * 1000 / df_brasil["População"]

# Calcular o PIB per capita
df_brasil.loc[:, "PIB per Capita (US$)"] = df_brasil["PIB (US$)"] / df_brasil["População"]

# Criar o gráfico de barras com dois eixos
fig, ax1 = plt.subplots(figsize=(10, 6))

# Gráfico de barras para o lixo por pessoa
ax1.bar(df_brasil["Ano"], df_brasil["Lixo por Pessoa (kg)"], color='green', alpha=0.7, label='Lixo por Pessoa (kg)')
ax1.set_xlabel('Ano')
ax1.set_ylabel('Lixo por Pessoa (kg)', color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Criar um segundo eixo y para o PIB per capita
ax2 = ax1.twinx()
ax2.plot(df_brasil["Ano"], df_brasil["PIB per Capita (US$)"], color='blue', marker='o', label='PIB per Capita (US$)')
ax2.set_ylabel('PIB per Capita (US$)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Título do gráfico
plt.title('Lixo Médio Gerado por Pessoa e PIB per Capita no Brasil (2014-2024)')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

# Paises mais populosos do dataset

```python
# Filtrar os dados para o ano de 2024
df_2024 = df[df['Ano'] == 2024]

# Agrupar os dados por país e pegar a população
populacao_por_pais = df_2024.groupby('País')['População'].max().reset_index()  # Usando max() para pegar a maior população por país
populacao_por_pais = populacao_por_pais.sort_values(by='População', ascending=False).head(10)  # Pegar os 10 mais populosos

# Criar o gráfico
plt.figure(figsize=(12, 6))
plt.barh(populacao_por_pais['País'], populacao_por_pais['População'], color='skyblue')
plt.xlabel('População')
plt.title('10 Países Mais Populosos em 2024')
plt.gca().invert_yaxis()  # Inverter o eixo y para que o país mais populoso fique no topo
plt.show()
```

#  Quantidade total de lixo gerado por cada país, TOP 10.

```python
import matplotlib.pyplot as plt

# Agrupar e somar o total de lixo por país
total_lixo_por_pais = df.groupby('País')['Total Lixo (Ton)'].sum()

# Obter os 10 países com o maior total de lixo
top_10_paises = total_lixo_por_pais.nlargest(10)

# Criar o gráfico de barras
top_10_paises.plot(kind='bar')
plt.title('Total de Lixo por País (Top 10)')
plt.ylabel('Total de Lixo (Ton)')
plt.xlabel('País')
plt.xticks(rotation=45)
plt.show()
```

# Analise a evolução da geração de lixo ao longo dos anos 

```python

# Calcular o total de lixo por país
total_lixo_por_pais = df.groupby('País')['Total Lixo (Ton)'].sum()

# Selecionar os 10 países com maior total de lixo
top_10_paises = total_lixo_por_pais.nlargest(10).index

# Criar o gráfico para os 10 países mais lixo
plt.figure(figsize=(12, 6))

for pais in top_10_paises:
    subset = df[df['País'] == pais]
    plt.plot(subset['Ano'], subset['Total Lixo (Ton)'], label=pais)
    
plt.title('Tendência de Lixo ao Longo dos Anos (Top 10 Países)')
plt.xlabel('Ano')
plt.ylabel('Total de Lixo (Ton)')
plt.legend()
plt.show()
```

- Novas colunas
- Aumento no Total de Lixo: Se o total de lixo gerado aumentou em relação ao ano anterior.
- Meta de Reciclagem: Se o país reciclou mais de 40% do lixo total.
- Aumento Populacional: Se a população aumentou em relação ao ano anterior.
- Aumento do PIB: Se o PIB aumentou em relação ao ano anterior.
