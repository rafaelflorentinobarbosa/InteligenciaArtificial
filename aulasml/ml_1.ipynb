# Situao Problema

| Imagem | Galinha_Papagaio | Tem Bico | Risco de Extino | Faz Cocorico |
|--------|------------------|----------|-------------------|--------------|
|     | 1                | Sim      | No               | Sim          |
|        | 0                | Sim      | Sim               | No          |
|        | 1                | Sim      | No               | Sim          |
|        | 0                | Sim      | Sim               | No          |
|        | 0                | Sim      | No               | No          |
|        | 1                | Sim      | Sim               | Sim          |

```python
# Papagaio

# Features [1 sim, 0 no] 
# Tem bico ?
# Risco de extinso ?
# Faz cocoric ?


papagaio1=[1,0,1]
papagaio2=[1,1,0]
papagaio3=[1,1,1]
```

```python
# Galinha

# Features [1 sim, 0 no] 
# Tem bico ?
# Risco de extinso ?
# Faz cocoric ?

galinha1=[1,0,1]
galinha2=[1,0,1]
galinha3=[1,1,1]
```

```python
# Dados
dados =[papagaio1,papagaio2,papagaio3,galinha1,galinha2,galinha3]
```

```python
# Classes
classes=[1,1,1,0,0,0]
```

```python
# Mostrar Dados
dados
```

# Treinamento do Modelo

```python
# Treinamento do Modelo
! pip install scikit-learn
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
```

```python
modelo.fit(dados,classes)
```

# Prever um animal

```python

# Features [1 sim, 0 no] 
# Tem bico ?
# Risco de extinso ?
# Faz cocoric ?

# tem bico, est em risco, faz no faz cocoric

animal_misterioso= [1,1,0]
previsao = modelo.predict([animal_misterioso])
```

```python
previsao
```

# Calculando a Acurcia do modelo
# Dados dos Rtulos

```python
# dados do treino x
treino_x = [papagaio1,papagaio2,papagaio3,galinha1,galinha2,galinha3]

# rotulos (classes) de treino y
treino_y =[1, 1, 1, 0, 0, 0]
```

```python
# animais misteriosos - para teste
misterioso1 =[1,0,1]
misterioso2 =[0,1,0]
misterioso3 =[0,0,1]
```

```python
# previsoes - dados de testes 
teste_x = [misterioso1,misterioso2,misterioso3]

# prevendo os testes
previsoes = modelo.predict(teste_x)
```

```python
# rotulos (classes)  de teste (avaliao- aqui e a sua provinha)
teste_y = [0,1,1]
```

```python
# importando o mdulo que j traz  a funo de acurcia encapsulada
from sklearn.metrics import accuracy_score

# definindo uma variavel para receber o retorno  da funo acurcia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100

# mostrando a acuracia
print(f"Acurcia: {taxa_de_acerto:.2f}%")
```
