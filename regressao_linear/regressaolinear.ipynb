## Qual o problema:

Preve preços de imóveis a partir de variáveis independentes.

Variáveis Disponíveis



1.  **area_primeiro_andar**: Refere-se à área do primeiro andar da propriedade, medida em metros quadrados.
2.  **existe_segundo_andar**: Esta variável é binária, indicando se a propriedade possui ou não um segundo andar. Pode ser representada como 1 para "sim" e 0 para "não".
3.  **area_segundo_andar**: Se a propriedade tiver um segundo andar, esta variável representa a área total do segundo andar, medida em metros quadrados.
4.  **quantidade_banheiros**: Indica o número total de banheiros na propriedade.
5.  **capacidade_carros_garagem**: Esta variável indica a capacidade da garagem da propriedade, ou seja, o número máximo de carros que podem ser estacionados na garagem.
6.  **qualidade_da_cozinha_Excelente**: Esta é uma variável categórica que avalia a qualidade da cozinha na propriedade. Neste caso, assume-se que se a cozinha for considerada "Excelente" é representada por 1, e caso contrário, por 0.
7. **preco_de_venda**: Este é o preço de venda da propriedade em reais. É a variável alvo que se tenta prever usando os outros atributos da propriedade.

```python
# podemos usar correlação
corr = df.corr()
```

```python
# intensidade - magnitude (metrica)
# visualizando correlacoes
corr['preco_de_venda']
```

```python
# para cada uma das caracteristicas tem uma ligacao com o preco da venda
# intensidade e direcao = -1 e 1
# hipotese o tamanho do primeiro andar pode estar correlacionado
```

1. Intensidade: refere-se à força da relação entre as variáveis medidas. Pode variar de -1 a +1, onde -1 indica uma correlação perfeitamente negativa, +1 indica uma correlação perfeitamente positiva e 0 indica ausência de correlação linear.
2. Direção: refere-se à natureza da relação entre as variáveis medidas. Uma correlação positiva indica que as variáveis aumentam ou diminuem juntas, enquanto uma correlação negativa indica que uma variável aumenta enquanto a outra diminui, ou vice-versa.
3. Exemplo: O coeficiente de correlação linear entre a área do primeiro andar e o preço de venda é de 0.62, indicando uma correlação positiva moderadamente forte (intensidade). Em geral, à medida que a área do primeiro andar de uma casa aumenta, o preço de venda também tende a aumentar (direção).

# Dividir treino e teste

O conjunto de treinamento 'e usado para ajustar o modelo, enquanto o conjunto de teste [e usado para avaliar seu desempenho em prever

```python
# import train_test_split
from sklearn.model_selection import train_test_split
```

```python
# definindo y e x
y = df['preco_de_venda'] # variavel resposta
x = df.drop(columns='preco_de_venda')
```

```python
# aplicando o split do y e x
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=230)
```

```python
# dados de treino para usar a formula
df_train = pd.DataFrame(data=X_train)
df_train['preco_de_venda']  = y_train
```

```python
# import ols
from statsmodels.formula.api import ols # essa biblioteca vai nos ajudar a criar um modelo de regressao
```

```python
# ajustando o primeiro modelo
modelo_0 = ols('preco_de_venda ~ area_primeiro_andar ',data=df_train).fit()
```

Conclusão

Dividimos nosso conjunto de dados, para treino e teste. Assim, nosso modelo consegue generalizar, ele aprende sobre diversas áreas.

```python
# vamos conhecer o modelo que a gente treinou
# para saber o efeito temos que aprender a interpretar os parametros da regressao linear

#visualizando parâmetros
modelo_0.params
```

area_primeiro_andar esta em reais, quando a gente aumenta um metro quadrado ele aumenta 6793 reais.
E se a area for zero.
Intercept - a gente espera que a casa puxe 152909
Quando o sinal esta positivo
Temos uma relacao linear positiva

```python
# existe um efeito significativo:
print(modelo_0.summary())
```

```python
# observando o R²
modelo_0.rsquared
```

```python
# obtendo resíduos
modelo_0.resid
```

```python
# mostrar em um histograma

modelo_0.resid.hist()
plt.title("Distribuição dos resíduos")
plt.show()
```

```python
# o outro pedaço de teste
# comparar que o modelo preveu e o valor realmente vendido
# obtendo o R² da previsão

y_predict = modelo_0.predict(X_test)

# importando o r2_score
from sklearn.metrics import r2_score
# printando r²

print("R²",r2_score(y_test,y_predict))
```
